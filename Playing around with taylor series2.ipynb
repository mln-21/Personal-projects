### Taylor Series of $\frac{1}{1-x}$

It can be expressed as $\sum_{n = 0}^{\infty}x^{n}$

```python
#import stuff
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-2, 2, 2000)
def exact_func(x):
    '''
    this is the function of 1/(1-x)
    '''
    return 1/(1-x)



def taylor_func(x, n = 10):
    '''
    taylor series of 1/(1-x) first given an x input or array and spits out an array, specify how many terms by setting 
    n = ... by defult n is 10
    '''
    n_terms = n
    c = np.zeros(len(x))
    for i in range(n_terms + 1):
        c += x**i
    return c, n

n = 30

fig, ax = plt.subplots(nrows = 3 , ncols = 1, figsize = (8,12))
ax[0].plot(x, exact_func(x))
ax[0].set_ylim(0,10)
ax[0].set_ylabel('Y')
ax[0].set_xlabel('X')
ax[0].set_title(f'graph of taylor series of 1/(1-x)')

ax[1].plot(x,taylor_func(x, n)[0])
ax[1].set_ylim(0,10)
ax[1].set_title(f'graph of taylor series of 1/(1-x) to n = {n} terms')
ax[1].set_ylabel('Y')
ax[1].set_xlabel('X')

ax[2].plot(x,abs(taylor_func(x, 20)[0] - exact_func(x)))
ax[2].set_ylim(0,10)
ax[2].set_title(f'graph of error of taylor series of 1/(1-x) to n = {n} terms')
ax[2].set_ylabel('Y')
ax[2].set_xlabel('X')

for a in ax:
    a.grid(True)
plt.tight_layout()
```

### Taylor Series of $e^{x}$

It can be expressed as $\sum_{n = 0}^{\infty}\frac{x^{n}}{n!}$

```python
x = np.linspace(-10, 10, 2000)
def exact_func(x):
    '''
    this is the function of 1/(1-x)
    '''
    return np.exp(x)


def factorial(i):
    n = 1
    for num in range(round(i)):
        n = n*(num+1)
    return n
def taylor_func(x, n = 10):
    '''
    taylor series of e^x first given an x input or array and spits out an array, specify how many terms by setting 
    n = ... by defult n is 10
    '''
    n_terms = n
    c = np.zeros(len(x))
    for i in range(n_terms + 1):
        c = c + x**i/(factorial(i))
    return c

terms = 30
fig, ax = plt.subplots(nrows = 3 , ncols = 1, figsize = (8,12))
ax[0].plot(x, exact_func(x))
ax[0].set_ylim(0,10)
ax[0].set_ylabel('Y')
ax[0].set_xlabel('X')
ax[0].set_title(f'graph o e^x')

ax[1].plot(x,taylor_func(x, n = terms))
ax[1].set_ylim(0,10)
ax[1].set_title(f'graph of taylor series of e^x to n = {terms} terms')
ax[1].set_ylabel('Y')
ax[1].set_xlabel('X')

ax[2].plot(x,abs(taylor_func(x, n = terms) - exact_func(x)))
ax[2].set_ylim(0,1)
ax[2].set_title(f'graph of error of taylor series of e^x to n = {terms} terms')
ax[2].set_ylabel('Y')
ax[2].set_xlabel('X')

for a in ax:
    a.grid(True)
plt.tight_layout()
plt.show()
```

## Let's try to make a general function for a Maclaurin series of any function that is continuous at x = 0

We will need symbolic imports to effectively take derivatives

Defining the function

```python
#note this is a much less efficent method of calculating maclaurin series
def maclaurin(func, terms = 10):
    '''
    imput a function of x and get back a maclaurin series
    '''
    def factorial(i):
        n = 1
        for num in range(round(i)):
            n = n*(num+1)
        return n
    
    mac = 0
    x = smp.symbols('x', real = True)
    for n in range(terms + 1):
        dfdx = smp.diff(func, x, n)
        dfdx_func = dfdx.evalf(subs={x: 0})
        mac = mac + dfdx_func*x**n/factorial(n)
    return mac
func = smp.exp(-5*x)*smp.sin(x)

print(maclaurin(func, 20))
t1 = %timeit -n10 -r3 -o maclaurin(func)
```

```python
def maclaurin(func, terms = 10):
    '''
    imput a function of x and get back a maclaurin series
    '''
    def factorial(i):
        n = 1
        for num in range(round(i)):
            n = n*(num+1)
        return n
    
    mac = 0
    fx = func
    x = smp.symbols('x', real = True)
    for n in range(terms + 1):
        dfdx_func = fx.evalf(subs={x: 0})
        mac = mac + dfdx_func*x**n/factorial(n)
        fx = smp.diff(fx, x)
    return mac
func = smp.exp(-5*x)*smp.sin(x)

print(maclaurin(func, 20))
t1 = %timeit -n10 -r3 -o maclaurin(func)
```

find a function we can take a Maclaurin series of

```python
x, a, b, c = smp.symbols('x a b c', real = True)

#defining the function with how many terms in the series
func = smp.log(1+x)
terms = 100
exact_f = smp.lambdify((x), func, "numpy")
fx = maclaurin(func, terms)
f = smp.lambdify((x), fx, "numpy")

# domain of the function
t = np.linspace(-0.99, 1, 1000)
y = f(t)
y_exact = exact_f(t)
```

### plotting the series vs the function

```python
def plot_it(fx, tspan, n = 1):
    x = smp.symbols('x', real = True)

    #defining the function with how many terms in the series
    func = fx
    terms = n
    exact_f = smp.lambdify((x), func, "numpy")
    fx = maclaurin(func, terms)
    f = smp.lambdify((x), fx, "numpy")

    # domain of the function
    t0, tf = tspan
    t = np.linspace(t0, tf, 5000)
    y = f(t)
    y_exact = exact_f(t)
    fig, ax = plt.subplots(nrows = 2 , ncols = 1, figsize = (8,12))
    ax[0].plot(t,y_exact , color = 'red' , label = f'{func}')
    ax[0].set_ylabel('y')
    ax[0].set_xlabel('t')
    ax[0].set_title(f'graph of {func}')

    ax[0].plot(t,y, color = 'black', label = 'Taylor series')
    ax[0].set_title(f'graph of Maclaurin series of {func} to n = {terms} terms')
    ax[0].set_ylabel('y')
    ax[0].set_xlabel('t')

    ax[1].plot(t,abs(y - y_exact), label = 'residual or error')
    ax[1].set_title(f'graph of error of Maclaurin series of {func} to n = {terms} terms')
    ax[1].set_ylabel('y')
    ax[1].set_xlabel('t')

    for a in ax:
        a.grid(True)
        a.legend()
    plt.tight_layout()
    plt.show()
```

## Eventually Taylor Series...

```python
def factorial(i):
    n = 1
    for num in range(round(i)):
        n = n*(num+1)
    return n
def taylor(func, terms, a = 0):
    '''
    taylor series centered at a, by default a = 0
    '''
    tay = 0
    fx = func
    x = smp.symbols('x', real = True)
    for n in range(terms + 1):
        dfdx_func = fx.evalf(subs={x: a})
        tay = tay + dfdx_func*(x-a)**n/factorial(n)
        fx = smp.diff(fx, x)
    return tay
```

```python
func = smp.exp(-5*x)*smp.sin(x)
taylor(func, 10, 1) 
```

update plot it function

```python
def plot_it(fx, tspan, n = 1, a = 0):
    x = smp.symbols('x', real = True)

    #defining the function with how many terms in the series
    func = fx
    terms = n
    exact_f = smp.lambdify((x), func, "numpy")
    fx = taylor(func, terms, a)
    f = smp.lambdify((x), fx, "numpy")

    # domain of the function
    t0, tf = tspan
    t = np.linspace(t0, tf, int((tf-t0)*500))
    y = f(t)
    y_exact = exact_f(t)
    fig, ax = plt.subplots(nrows = 2 , ncols = 1, figsize = (8,12))
    ax[0].plot(t,y_exact , color = 'red' , label = f'{func}')
    ax[0].set_ylabel('y')
    ax[0].set_xlabel('t')
    ax[0].set_title(f'graph of {func}')

    ax[0].plot(t,y, color = 'black', label = 'Taylor series')
    ax[0].set_title(f'graph of Taylor series of {func} centered at a = {a} to n = {terms} terms')
    ax[0].set_ylabel('y')
    ax[0].set_xlabel('t')

    ax[1].plot(t,y - y_exact, label = 'residual or error')
    ax[1].set_title(f'graph of error of Taylor series of {func} centered at a = {a} to n = {terms} terms')
    ax[1].set_ylabel('y')
    ax[1].set_xlabel('t')

    for a in ax:
        a.grid(True)
        a.legend()
    plt.tight_layout()
    plt.show()
```

```python
plot_it(smp.exp(x), [0, 10], n = 15, a = 4)
```
