## Population size iteration
 Based on the formula
 <center>$N(x) = xe^{r(1 - \frac{x}{k})} $

```python
import numpy as np
import matplotlib.pyplot as plt

def func(x, r = 1/2, k = 10000):
    return x*np.exp(r*(1-x/k))
def dfunc(x, r = 1/2, k = 10000):
    return np.exp(r*(1-x/k))*(1-x*r/k)

def popsize(n,IC = 1000,r = 1/2, k = 10_000):
    x0 = IC
    x_list = np.zeros(n)
    x_list[0] = x0
    dx_list = np.zeros(n)
    dx_list[0] = dfunc(x0, r , k)
    
    for i in range(n - 1):
        x_list[i + 1] = func(x_list[i], r, k)
        dx_list[i + 1] = dfunc(x_list[i + 1], r , k)
    
    plt.figure(figsize = (8, 8))
    plt.subplot(211)
    plt.scatter(range(n),x_list, label = 'Population size', s = 5, marker = 'x')
    plt.grid(True)
    plt.title(f'Population over {n} generations')
    plt.ylabel('Population size')
    plt.xlabel('n generations')
    plt.legend()
    
    plt.subplot(212)
    plt.tight_layout()
    plt.scatter(range(n),dx_list, s = 5, marker = 'o')
    plt.grid(True)
    plt.title(f'Rate of population growth over {n} generations')
    plt.ylabel('Rate of population growth')
    plt.xlabel('n generations')
    plt.tight_layout()

    return x_list[-1]


starting_population = 0.5
popsize(200, starting_population, r = 1 , k = 10000)
```
